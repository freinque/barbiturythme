## Example: use BarBitURythme to encode drum rhythms

<p> This example illustrates how BarBitURythme can be used to</br>
<ul>
    <li> <b>encode</b> rhythm sequences of a signal.</li>
</ul>
</p>
<p>
    That encoding will later be used to <a href="https://github.com/freinque/barbiturythme/html/bbur_gener_test_0.0.4.html"><span>train BarBitURythme statistical models</span></a> and generate sequences of similar style.
</p>

We set up the python machinery that will be used. Since barbiturythme is not in our system's python package repositories (sys.path), we add it.

```python
import os, sys
sys.path.append( '/'.join(os.getcwd().split('/')[:-1]) + '/barbiturythme' )
```

```python
import numpy as np
import subprocess
import barbiturythme
from barbiturythme import bbur_gene
```

We choose the data folder/files location and name and initialize a <b>bbur_gene.bbur_io</b> class to read and write to data files. 
<ul>
    <li><b>io.data_folder_name</b> is the name of a folder in ../data</li>
    <li><b>io.data_file_name</b> is the name of the files in ../data/data_folder_name</li>
    <li><b>io.n_freq_int</b> is the number of percussive instruments encoded in those files, e.g.
    Roland-808 kick drum, snare and hihat </li>
</ul>

```python
io = bbur_gene.bbur_io()
io.data_folder_name = '4bars_4beats_4subs_3bands'
io.data_file_name = 'hh_00' 
io.n_freq_int = 3
```

Then the class <b>barbiturythme.BarBit</b> will be used to read .wav files and encode (some of) their content into machine adapted data.

```python
bb = barbiturythme.BarBit()
bb.read( '/home/freinque/tracks/atlanta_trap.wav' )
```

We set up the parameters of the standard spectral analysis procedure. Roughly speaking, we will 
<ul>
    <li>scan the signal using a 'magnifying glasses' of (time) width <b>bb.window_size</b>, making steps of (time) length <b>bb.hop_size</b>,</li>
    <li>look into 'magnifying glasses' with filters that only let the frequencies in the intervals <b>bb.freq_bounds</b> pass, and</li>
    <li> calculate the increase of intensity in each of them from one step to the next with <b>bb.calc_odf()</b>.</li>
</ul>
Here, we use a frequency interval [20,120] to get the bass drum (this seems standard), [4000,6000] to catch the Roland-808 snare and [13000,14000] to see the hihat. If you don't know what are the main frequencies of a percussive instrument, I suggest a Fourier transform (over a tight window containing a hit of the instrument) using the Analyse > Plot Spectrum tool in Audacity.

```python
bb.window_size = 1024*2
bb.hop_size = 512
bb.set_freq_bounds([[20,120],[4000,6000],[13000,14000]])
bb.smooth_windows = [1,10,1]
start_time = 10
duration = 15
bb.calc_odf( start_time, start_time + duration )
```

We plot the increase levels in each interval

```python
bb.plot_odf()
```

We adjust the levels at which we consider that each instrument is played and use <b>bb.calc_onsets()</b> to determine at which steps each instrument is played.

```python
bb.min_thresholds = [0.3,0.5,0.2]
bb.median_windows = [5,9,13]
bb.peak_sizes = [0.05,1,0.02]
bb.calc_onsets()
```

We plot the results. If they don't make us happy (i.e. what we hear and see coincide), we can reexecute the above cells with different paramenters.

```python
bb.plot_onsets()
```

At this point, our BarBit doesn't know what is the tempo of our track so it can't really find its bars (i.e. measures) or multi-bar structures (e.g. choruses, etc.). If you know it for sure, just set <b>bb.bpm</b> to it.

Otherwise, <b>bbur_gene.bpm_finder()</b> provides a decent (although very simple) way of calculating it. Here, we use the onsets (i.e. moments when hit) of the snare and multiply the result by 2 since it is played every two beats.

```python
bpmf = bbur_gene.bpm_finder()
bpmf.gcd = bb.hop_size
bpmf.sampling_rate = bb.sampling_rate
bpmf.set_onsets( bb.onsets[1] )
```

```python
bpmf.calc_bpms()
```

```python
bb.bpm = 2.*bpmf.bpms[3]
bb.bpm
```

We set the set the parameters of the bar encoding.
<ul>
    <li> <b>bb.beats_per_bar</b> is the time signature of the track,</li>
    <li> <b>bb.n</b> is the length of the sequences (in bars) that we want to encode,</li>
    <li> <b>bb.subs_per_beat</b> is the time resolution that we want to use in the encoding (i.e. number of subdivisions per beat).</li>
</ul>

```python
bb.subs_per_beat = 4
bb.beats_per_bar = 4
bb.n = 4
```

<b>bb.encode_onsets()</b> then does what its name says.

```python
bb.encode_onsets()
```

The result is a bunch of 4bars (concatenated sequence made of 4 bars) in integer (<b>bb.n_bars_int</b>) and binary (<b>bb.n_bars_bin</b>) forms. For instance

```python
bb.n_bars_bin[0][0]
```

where the third element being [1,0,1] means that at the third time subdivision, the fisrt and third (kick and hihat) intruments are played, the the fourth element [0,0,1] means that at the fourth time subdivision, the third (hihat) intrument is played, etc. The integer form looks like

```python
bb.n_bars_int[0][0]
```

where 5 stands for 1*2**0 + 0*2**1 + 1*2**2, etc.

bb doesn't know where exactly to set the 'time 0' moment. We help a bit, but this could be mostly automated. <b>subdiv_shift</b> is our preferred time shift.

```python
subdiv_shift = int(( bb.sampling_rate*60.0/(bb.bpm*bb.hop_size) ) / bb.subs_per_beat) #time shift, in subdivs
observations = np.array( bb.n_bars_int[57*subdiv_shift][:] )
```

```python
print observations[:]
```

And we add de desired 4bars to the data file.

```python
io.append_data( [observations[1]] )
```
